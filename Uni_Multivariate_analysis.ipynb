```python
def show_risk_normal(df,var,key):
    try:

        keys = list(df[var].unique())
        print(keys)
        totals_df = Counter(df[var]) 
        
        claims_df = Counter(df.query("claimed_on>0")[var])
#         .query("{}=='{}'and claimed_on>0 ".format('gender','Female')).claimed_on.count()
        
        claims = df.query("{}=='{}' and claimed_on>0".format(var,keys[key])).claimed_on.sum()
        
        claim_exposure_sum = df.query("{}=='{}'".format(var,keys[key])).policy_on.sum()
        
        return print("total_counts:") ,\
               display(pd.DataFrame.from_dict(totals_df, orient='index').reset_index().rename(columns={'index':var, 0:'count'})) ,\
               print("claim_counts:") ,\
               display(pd.DataFrame.from_dict(claims_df, orient='index').reset_index().rename(columns={'index':var, 0:'count'})) ,\
               print("month_exposure_sum:" , claim_exposure_sum ) ,\
               print(f"{keys[key]} risk: claims/exposure:" , round((claims/claim_exposure_sum)*100,1),'%' ) 
    except Exception as e:
        return (e, " no such category")
```
